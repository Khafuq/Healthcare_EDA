```python
X1 = df.groupby("Gender").agg({"Blood Type": lambda x: ', '.join(x), "Age": 'mean'}).reset_index()

print(X1)
```

```python
X1 = df.groupby("Gender").agg({"Blood Type": lambda x: ', '.join(x), "Age": 'mean'}).reset_index()

plt.figure(figsize=(9, 6))
sns.barplot(x='Blood Type', y='Age', hue='Gender', data=df, ci=None, palette="Set1")
plt.title("Mean Age by Gender and Blood Type")
plt.xlabel("Blood Type")
plt.ylabel("Mean Age")
plt.legend(title='Gender')
plt.show()
```

```python
x2 = df.groupby(['Medical Condition', 'Test Results'])[['Billing Amount']].sum().reset_index()

plt.figure(figsize=(9,6))
sns.barplot(x= x2['Medical Condition'], y= x2['Billing Amount'], hue=x2['Test Results'], ci=None, palette="Set2")
plt.title("Billing Amount according to Medical Condition and Test Results")
plt.ylabel("Billing Amount")
plt.xticks(rotation=45, fontsize = 9)
plt.show()
```

```python
x2 = df.groupby(['Medication'])[['Billing Amount']].sum().reset_index()

plt.figure(figsize=(9,6))
sns.barplot(x= x2['Medication'], y= x2['Billing Amount'], ci=None, palette="Set2")
plt.title("Billing Amount according to Medical Condition and Test Results")
plt.ylabel("Billing Amount")
plt.xticks(rotation=45, fontsize = 9)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 1, figsize=(7, 7))

# Gender
axs.pie(df['Gender'].value_counts(), labels=df['Gender'].value_counts().index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
axs.set_title('Gender Distribution')

plt.show()
```

```python
fig, axs = plt.subplots(1,figsize=(7, 20))
#  Blood Type
axs.pie(df['Blood Type'].value_counts(), labels=df['Blood Type'].value_counts().index, autopct='%1.1f%%', colors=['lightgreen', 'salmon', 'skyblue'])
axs.set_title('Blood Type Distribution')

plt.show()
```

```python
fig, axs = plt.subplots(1,figsize=(7, 5))

axs.hist(df['Test Results'], color='salmon')
axs.set_title('Test Results Distribution')
axs.set_xlabel('Test Results')
axs.set_ylabel('Frequency')

plt.show()
```
